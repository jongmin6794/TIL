```python
titanic=sns.load_dataset("titanic")
titanic.info()
```

```python
def count_missing(vec):
    #누락값의 개수 리턴
    nv=pd.isnull(vec)
    nc=np.sum(nv)#true:1,false:0
    return nc
    #누락값의 개수 리턴
    
# def count_missing(vec):
#     return vec.isnull().sum(axis=0)

def prop_missing(vec):
    num=vec.size
    miss=count_missing(vec)
    return miss/num
    
def prop_complete(vec):
    return 1-prop_missing(vec)
```

```python
cm=titanic.apply(count_missing)
print(cm)
#각 컬럼별 누락값의 개수가 모두 출력
#각 컬럼별 누락값의 비율 모두 출력
pm=titanic.apply(prop_missing)
print(pm)

pcm=titanic.apply(prop_complete)
print(pcm)
```

```python
#행 방향으로 누락값의 개수/비율 조사
cmr=titanic.apply(count_missing, axis=1)
pmr=titanic.apply(prop_missing, axis=1)
pcmr=titanic.apply(prop_complete, axis=1)
cmr.head()
pmr.head()
pcmr.head()
```

```python
titanic['num_missing']=titanic.apply(count_missing, axis=1)
titanic
```

```python
#[1880df, 1881df, ..., 2010df]
pieces=[]
for year in range(1880,2011):
    path="C:/Users/student/Downloads/names/names/yob%d.txt" %year
#     print(path)
    df=pd.read_csv(path,names=['name','sex','births'])
    df['year']=year
    pieces.append(df)
# pieces
# pd.concat(pieces).shape
names=pd.concat(pieces, ignore_index=True)
```

```python
# birth를 데이터, 행을 year, 열을 sex 피봇테이블 생성
names.pivot_table(values='births', index='year',columns='sex',
                  aggfunc=np.sum)
#기본설정은 평균! aggfunc=np.mean이 디폴트
total_births=pd.pivot_table(data=names,values='births',index='year',columns='sex'
                           ,aggfunc=sum)
total_births.tail()
```

```python
# names에서 year, sex
# ex) mary는 1880년에 태어난여자아이들 중 
# 전체에서 차지하는 비중이 얼마인지 조사?
def add_prop(gruop):
    #구현
    gruop['prop']=gruop.births/gruop.births.sum()
    return gruop
names=names.groupby(['year','sex']).apply(add_prop)
names
names.head(20)
```

```python
np.allclose(names.groupby(['year','sex']).prop.sum(),1)
#나온 값이 잘 나왔는지 검산할때 주로 사용
```

```python
#각 연도별, 성별에 따른 이름 빈도수가 가장
#높은 이름을 1000개씩 추출
def get_top1000(group):
    return group.sort_index(by='births',ascending=False)[:1000]

grouped=names.groupby(['year','sex'])
top1000=grouped.apply(get_top1000)
#그룹 인덱스 제거
top1000.reset_index(inplace=True, drop=True)
top1000
```

```python
def get_qc(group,q=0.5):
    group=group.sort_index(by='prop',ascending=False)
    return group.prop.cumsum().values.searchsorted(q)
    
res=top1000.groupby(['year','sex']).apply(get_qc)
res=res.unstack('sex')
#unstack을 안하고 출력하면 합쳐져 나옴...?
#res를 출력
res.plot()
```
