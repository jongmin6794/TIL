```python
def seq2dataset(seq, windowSize):
    dataset=[]
    for i in range(len(seq)-windowSize):#54-4=50
        subset=seq[i:(i+windowSize+1)]
        dataset.append([code2idx[item] for item in subset])
    return np.array(dataset)
```

```python
print(seq2dataset(seq,windowSize=4))
```

```python
dataset=seq2dataset(seq,windowSize=4)
dataset.shape
```

```python
xTrain=dataset[:,0:4] # 나 비 야 나, ...
yTrain=dataset[:,4] # 비, ...
maxIdxValue=13
```

```python
#입력값 정규화
xTrain=xTrain/maxIdxValue
xTrain
```

```python
from keras.utils import np_utils
yTrain=np_utils.to_categorical(yTrain)
yTrain.shape
#(50,)->(50,12)
```

```python
oneHotVecSize=yTrain.shape[1]#y:12가지 음계
```

```python
from keras.models import Sequential
from keras.layers import Dense
import keras
```

```python
#모델 생성
model=Sequential()
model.add(Dense(128, input_dim=4, activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(oneHotVecSize, activation='softmax'))
```

```python
model.summary()
# 128 * 4 + 128 =640
# 128 * 128 + 128 = 16512
# 128 * 12 + 128 = 1548
```

```python
#모델 학습과정
model.compile(loss="categorical_crossentropy",optimizer='adam',
             metrics=['accuracy'])
```

```python
class LossHistory(keras.callbacks.Callback):
    def init(self):
        self.losses=[]
        print("호출됨")
    def onEpochEnd(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
```

```python
history=LossHistory()
history.init()
```

```python
model.fit(xTrain,yTrain,epochs=2000,batch_size=10,callbacks=[history])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.logissess)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python
scores=model.evaluate(xTrain,yTrain)
print("%s:%.2f%%" %(model.metrics_names[1],scores[1]*100))
```

```python
from keras.layers import LSTM
model=Sequential()
model.add(LSTM(128, input_shape=(4,1)))
#128출력셀, 4타임스텝, 1 속성개수
#stateful=기본값이 False, 여기서는 True로 해주어야 함.
model.add(Dense(oneHotVecSize, activation='softmax'))
"""
타임스텝(windwo size):샘플 1개에 포함된 시퀀스 개수(input size)
속성:입력되는 음표 1개당 index 1개 입력

"""
```

```python
np.shape(xTrain)#샘플수, 타임스탭, 속성수
xTrain=np.reshape(xTrain,(50,4,1))
```

```python
#모델 학습과정
model.compile(loss="categorical_crossentropy",optimizer='adam',
             metrics=['accuracy'])
```

```python
model.fit(xTrain,yTrain,epochs=2000,batch_size=14,callbacks=[history])
```

```python
scores=model.evaluate(xTrain,yTrain)
print("%s:%.2f%%" %(model.metrics_names[1],scores[1]*100))
```

```python
## 앞에 일반 신경망보다 lstm을 썻는데도 성능이 더 떨어졌다.
# 왜그럴까? stateful? 상태유지모드 적용되지 않았기 때문
"""
LSTM(메모리셀개수, input_dim=입력송성수, input_length=시퀀스데이터입력길이)
return_sequences : 시퀀스출력여부
False=마지막 시퀀스에서 한번만 출력(many to one)
True=각 시퀀스에서 출력(many to many)

-stateful:상태유지모드
학습 샘플의 가장 마지막 상태가 다음 샘플
학습할때 입력으로 전달할 것인지 여부

"""
```

```python
"""
상태유지모드(stateful)에서 현재 샘플의 학습
상태가 다음 샘플의 초기 상태로 전달

LSTM 셀 내부적으로 기억할 것은 기억하고,
버릴 것은 버리도록 하기위한 옵션
stateful=True

상태유지모드에서는 입력형태를(batch_input_shape)
(배치사이즈, 타임스텝, 속성)으로 설정

"""
```

```python
model=Sequential()
model.add(LSTM(128, batch_input_shape=(1,4,1),stateful=True))
#128출력셀, 4타임스텝, 1 속성개수
#stateful=기본값이 False, 여기서는 True로 해주어야 함.
model.add(Dense(oneHotVecSize, activation='softmax'))
```

```python
"""
-한 에폭안에서 여러개의 시퀀스 데이터가 있을때,
새로운 시퀀스 데이터를 학습하기 전에 상태 초기화 필요

-마지막 샘플 학습을 마치고나서, 새로운 에폭이
실행될때에는 새로운 샘플 학습을 해야하므로 상태 초기화가 필요

"""
```

```python
#모델 학습과정
model.compile(loss="categorical_crossentropy",optimizer='adam',
             metrics=['accuracy'])
```

```python
numEpochs=2000
```

```python
for i in range(numEpochs):
    print("에폭:"+str(i))
    model.fit(xTrain,yTrain,epochs=1,batch_size=1,callbacks=[history],shuffle=False)
    model.reset_states()
    #state 정보를 초기화
```

```python
model.evaluate(xTrain,yTrain,batch_size=1)
print("%s:%.2f%%" %(model.metrics_names[1],scores[1]*100))
```
