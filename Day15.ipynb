```python
#데이터프레임을 인덱스를 기준으로 병합

dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']},
    index= ['K0', 'K1', 'K2', 'K3'])
dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
    index=['K2', 'K3', 'K4', 'K5'])
```

```python
#index 기준 병합 방법 : join(), merge()
#index 기준 left join 수행

#merge()를 이용한 left join
pd.merge(dfleft,dfright, left_index=True,right_index=True)
pd.merge(dfleft,dfright, left_index=True,right_index=True, how='right')
```

```python
pd.merge(dfleft,dfright, left_index=True,right_index=True, how='right')
#merge함수는 판다스가 가지고 있지만 join함수는 데이터프레임이
dfleft.join(dfright) #how=left
dfleft.join(dfright, how='right')
```

```python
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']},
    index= ['K0', 'K1', 'K2', 'K3'])
dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
    index=['K2', 'K3', 'K4', 'K5'])
```

```python
#index기준 합치기,(merge함수 이용)
#k0
#k1
#k2
#k3

pd.merge(dfleft,dfright,left_index=True,right_index=True,how='left')
#join함수 이용 동일한 결과로
dfleft.join(dfright,how='left')
dfright.join(dfleft,how='right')
#merge함수 이용 outer join
pd.merge(dfleft,dfright,left_index=True,right_index=True,how='outer')
#join함수로 outer
dfleft.join(dfright,how='outer')
```

```python
#index와 key를 조합하여 merge
dfleft = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})
dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
   index=['K2', 'K3', 'K4', 'K5'])
print(dfleft)
print(dfright)
pd.merge(dfleft,dfright,left_on='KEY',right_index=True, how='left')
dfleft.join(dfright,on='KEY',how='left')
```

```python
#문제6
import numpy as np
inputData=[0,0,3,5,2,4,9,0,6,4,0,6,0,0]
buildData=[[0]*14]*10
buildData=np.array(buildData)
count=0
for i in range (0,14):#그저 보기 편하게
    for j in range(0,inputData[i]):
        buildData[9-j][i]=1
print(buildData)
for i in range(2,12):#가장 높은 층만 찾아서 조망권 검사
    j=0
    while True:
        print(i,inputData[i]-j,inputData[i-1],inputData[i-2],inputData[i+1],inputData[i+2])
        if (inputData[i]-j) > (inputData[i-1]) and (inputData[i]-j) > (inputData[i-2]) and (inputData[i]-j) > (inputData[i+1]) and (inputData[i]-j) > (inputData[i+2]):
            count+=1
            j+=1
        else:
            break
print(count)
```
