```python
"""
np.digitize함수, np.where함수:연속형 변수 -> 이산화
이항변수화: 0또는 1 값을 갖는 변수를 만드는 것
이산(형)화: 연속형 변수를 2개 이상의 범주를 갖는 변수로 변환
ex)점수:75 -> A/B/C/D/F(범주=category)
"""
np.random.seed(10)#seed를 갖게 주면 같은 난수가 출력 42, 777, 1004
#모델 만드는 과정에서 난수를 많이사용(모든 데이터 사용x 그중 랜덤 일부 사용)
#여러 테스트 시, 데이터 다르지 않도록 하기 위함
np.random.randn(1)
```

```python
"""
df의 c1열 20개 data구성
10개 bin으로 균등하게 나눔
"""
df.C1.min() #-1.977
df.C1.max() # 1.484
bins=np.linspace(df.C1.min(), df.C1.max(),10)
bins
df.C1
df['C1_bin']=np.digitize(df['C1'],bins)
df
```

```python
#이산형화를 왜 많이 할까?
#구간별로 요약통계, 범주간 평균 차이, 독립성 검정
#indexing...
df.groupby('C1_bin')['C1'].size()
df.groupby('C1_bin')['C1'].mean()
df.groupby('C1_bin')['C1'].std()
df.groupby('C1_bin')['C2'].value_counts()

df[df['C1_bin']==4]
```

```python
#pd.get_dummies():가변수 생성
#겟 더미를 이용 온핫 인코딩
df.C1_bin
pd.get_dummies(df['C1_bin'],prefix='C1')
#모델=x1*가중치+x2*가중치+...+x3*가중치
```

```python
#np.where(조건,참,거짓)
df["h_l"]=np.where(df['C1']>=df.C1.mean(),"high","low")
df
df.groupby("h_l")['C1'].size()
df.groupby("h_l")['C1'].mean()
df.groupby("h_l")['C1'].std()
```

```python
q1=np.percentile(df['C1'],25)
q3=np.percentile(df['C1'],75)
#연습문제:C1열 값이 <= q1 이면 df['h_m_l']=ㅣ
#C1열 값이 > q1 and <=q3 df['h_m_l']=m
#else df['h_m_l']=h
df['h_m_l']=np.where(df['C1']>=q3,'high',np.where(df['C1']>=q1,
                                                 'medium','low'))
df
```
