```python
[]: [] 사이의 문자들과 매치.[A-K], [0-9], [^0-9]
정규식:[ABCDE]
문자열 A:매치, BLUE:매치, DRY:매치, SKY:매치안됨

2.
\d <=> [0-9], \D <-> [^0-9]
\w <=> [a-zA-Z0-9_], \W <=> [^a-zA-Z0-9_]

3..(dot) : 모든 문자
정규식 : a.b <=> a + 모든문자 + b
문자열 aab : 매치, abb : 매치

*(0번 이상 반복)
정규식 : ca*t
문자열 : ct 매치, cat :매치, caaaat : 매치
+(1번 이상 반복)
정규식 : ca+t <-> a가 1번 이상 반복
문자열 ct : 매치x, cat : 매치, caaaat : 매치
            
ca{2}t <=> c + a는 반드시 2번 반복 + t
문자열 cat : 매치x, caat : 매치
        
{2,5} : 2~5번 반복
ca{2,5}t <=> c+a는 2번이상 5번이하 + t
문자열 cat : 매치x, caaat : 매치
ca?t <=> c + a(있어도 없어도 됨) + t
```

```python
res=p.finditer("Life is too short")
for r in res:
    print("start()",r.start())#매치 시작 위치
    print("end()",r.end()) #매치 끝 위치
    print("group()",r.group())#매치 문자열
    print("span()",r.span())#(시작, 끝)
```

```python
#정규식 컴파일 옵션
p=re.compile("a.b") # .은 모든 문자와 매치(\n 문자 제외)
#m=p.match("acb")
m=p.match("a\nb")
print(m)

#DOTALL 옵션: .을 사용할 \n 문자도 포함하고자 하는 경우
p=re.compile("a.b",re.DOTALL)
m=p.match("a\nb")
print(m)

#re.I 옵션 : ignorecase는 대소문자 구분없이 수행
p=re.compile("[a-z]")
print(p.match("python"))
print(p.match("Python"))
p=re.compile("[a-z]",re.I) # "[a-z][A-Z]"과 같음
print(p.match("Python"))
```

```python
# MULTILINE 옵션
text="""python one
python two
you need python
"""
p=re.compile("^python\s\w+")
print(p.match(text))
print(p.search(text))
print(p.findall(text)) 
#^를 문자열 전체의 처음에 대해서 적용되기때문
# 그런데, 내가 원하는 것은 문자열 전체가 아니라
# 문자열의 각 라인 단위로 정규식 (^)를 적용 => MULTILINE

p=re.compile("^python\s\w+", re.MULTILINE)
print(p.findall(text))
```

```python
"""
kim 010-2345-6789
hong 02-1234-5678
lee seoul(x)
park 010 1234 5656(x)
"""

pat=re.compile("\w+\s+\d+[-]\d+[-]\d+")
m=pat.search("kim 010-2345-6789")
#m=pat.search("hong 02-1234-5678")
#m=pat.search("lee seoul")
#m=pat.search("park 010 1234 5656")
print(m)
```

```python
pat=re.compile("(\w+)\s+\d+[-]\d+[-]\d+")
m=pat.search("kim 010-2345-6789")
print(m.group())
#():그루핑 기호
# 이름 그루핑 => (\w+)
print(m.group(1)) #첫번째 그룹에 해당되는 문자열
#print(m.group(2)) 그룹이 없으므로 에러
print(m.group(0)) # 매칭된 전체 문자열
```

```python
pat=re.compile("(\w+)\s+(\d+)[-](\d+)[-](\d+)")
m=pat.search("kim 010-2345-6789")
print(m.group(0))
print(m.group(1))
print(m.group(2))
print(m.group(3))
print(m.group(4))
```

```python
pat=re.compile("(\w+)\s+(\d+)[-]((\d+)[-](\d+))")# 괄호 중첩일시 바깥족부터 번호 할당
m=pat.search("kim 010-2345-6789")
print(m.group(0))
print(m.group(1))
print(m.group(2))
print(m.group(3))
print(m.group(4))
print(m.group(5))
```

```python
#그루핑시 이름을 부여
#pat=re.compile("(\w+)\s+(\d+)[-]((\d+)[-](\d+))")
#이름 부여 : 형식 => (?P<그룹명>...)
#(\w+) => (?P<name>\w+)
pat=re.compile("(?P<name>\w+)\s+(\d+)[-]((\d+)[-](\d+))")
m=pat.search("kim 010-2345-6789")
print(m.group(0))
print(m.group(1))
print(m.group("name"))
```

```python
# 함수에서 '이름', '인수' 추출하는 패턴식
# ex)
# 함수 구문 : sumPrice(50)
# 함수명 : sumPrice, 인수 : 50
g=re.match("(?P<func>[a-z][a-zA-Z0-9_]+)\((?P<arg>[0-9]+)\)","sumPrice(50)")
m1=g.group("func") # sumPrice
m2=g.group("arg") # 50
print(m1)
print(m2)
```

```python
#2번문제 토지 파일 전처리(정규식) 
# 1. 저자명 추출
# 2. 편 제목 추출 ex) 제 2편 제목-추적과 음모
# 3.각 편에 대한 장 구성 정보 출력 ex) 제 1편 - ??장으로 구성
# 4. 인용부호(겹따옴표)로 묶여있는 대사만 추출 -> 챗봇만들기가능
# 5. 치환(양반, 사람 => 사람)

#<author>박경리</author>
#문제1 txt 파일 안의 일정 문자 고쳐 저장
import re
f=open("C:/Users/student/Downloads/Python_JP/BEXX0003.txt","r",encoding="UTF-16")
lines=f.readlines()
# pat = re.compile("양반|사람")
# pat.sub("사람",lines[i])
#[a-zA-Z0-9\s]+
jcount=0
acount=0
for i in range(len(lines)):
    pat = re.search("\<author\>(?P<author>[가-힣]+)",lines[i])
    ptitle = re.search("<head>(?P<Ptitle>제[0-9]+편\s[가-힣\s]+)",lines[i])
    jtitle = re.search("<head>(?P<Jtitle>[0-9]+장\s[가-힣\s]+)",lines[i])
    if pat and acount<1:
        print("저자명 :",pat.group("author"))#두번출력 싫은데...
        acount+=1
    if ptitle:#편제목 추출
        print("편제목 :",ptitle.group("Ptitle"))
        jcount=0
    if jtitle:
        print("  장제목 :",jtitle.group("Jtitle"),", 장 갯수 :",jcount+1,"개")
        jcount+=1
    if '양반' in lines[i]:
        lines[i]=lines[i].replace("양반","사람")
#print(lines)
tcount=0
for i in range(len(lines)):
    talk = re.search("(?P<talking>\"[a-zA-Z0-9가-힣-.,!?\s]*\")",lines[i])
    if talk:
        if tcount>4:
            print("인용부호:",talk.group("talking"),tcount-4)
        tcount+=1
f.close()
f=open("C:/Users/student/Downloads/Python_JP/BEXX0003.txt","w",encoding="UTF-16")
for i in range(len(lines)):
    f.write(lines[i])
f.close()
```
