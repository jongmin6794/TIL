```python
#문제2
bins=[5,20,30,40,60,90]
labels=['미성년자','청년','중년','장년','노년']
# train.groupby('Survived')['Age'].mean()
liveg=train[train['Survived']==1]#생존자 그룹
deadg=train[train['Survived']==0]#사망자 그룹
liveGn=liveg['Age'].dropna(axis=0)#생존자 그룹 나이 결측값 제거
liveGn
deadGn=deadg['Age'].dropna(axis=0)#사망자 그룹 나이 결측값 제거
deadGn
catsL=pd.cut(liveGn,bins,labels=labels)
catsD=pd.cut(deadGn,bins,labels=labels)
catsL
df=pd.DataFrame()
df['생존자']=catsL.value_counts()
df['생존자 비율']=catsL.value_counts()/catsL.value_counts().sum()
df['사망자']=catsD.value_counts()
df['사망자 비율']=catsD.value_counts()/catsD.value_counts().sum()
df
```

```python
#Microsoft R서버를 활용한 빅데이터 분석
import pandas as pd
import seaborn as sns
tips = sns.load_dataset("tips")
#1)
# tips.astype({'sex':'object'}).dtypes
tips.sex=tips['sex'].astype('str')#str이 저절로 object형식으로
tips.sex=tips['sex'].astype('object')
tips.info()
#2)
tips
tips.loc[[1,3,5,7],'tip']='missing'
tips
#3)scikitlearn에 있는 preprocessing 모듈중 표준화 및 정규화와 
#관련된 함수를 모두 기술하시오
#RobustScaler():이상치가 있는 데이터 표준화
#percentile():표준화 식에서 이상치가 있는 데이터를 표준화 하는 식에서 IQR구하는 데에 사용된다.
#median():표준화 식에서 이상치가 있는 데이터를 표준화 하는 식에서 평균대신 사용된다.
#np.std() : np는 넘파이, 표준화를 위한 표준편차를 구할때 사용한다.
#MinMaxScaler() : 정규화로 0~1사이의 값으로 변경해준다.
#minmax_scale() : MinMaxScaler()보다 약간 편의 기능이 추가된 같은 개념의 정규화 함수이다.
# 2.표준화와 정규화의 차이점을 설명하시오.
#표준화: ( 데이터-(각 열,행 데이터 평균) )/각 열,행 표준편차
#       평균을 기준으로 어느 정도 떨어져 있는지를 나타낼때 사용된다.
#       2개 이상의 대상이 단위가 다를 때 대상 데이터를 같은 기준으로 볼 수 있도록 한다.
#       예시로 몸무게표준화- 키표준화+이면 몸무게는 평균보다 마른편에 키는 큰 편의 사람이 된다.
#정규화: 0~1사이의 범위로 데이터를 표준화 특히 인공신경망에서 사용하기 용이하다.
#둘다 개념적으로 데이터 rescaling의 공통점을 가지지만, 수식이 다르고, 표준화는
#특정 범위를 벗어난 데이터는 outlier로 간주해 제거한다. 그러나 정규화는 min이 0에 max값이 1에
#가깝게 분포가 된다.또한 코딩적 측면에서 표준화는 1개 컬럼만 가능하지만 
#정규화는 여러 컬럼을 한꺼번에 정규화가 가능하다.

#3.데이터 분석을 위해 필요로 하는 라이브러리 및 용도를 기술하시오.
#pandas : 데이터프레임과 Series을 제공한다. 데이터 프레임은 일종의 엘셀과 같은
#형태의 스프레드시트형태의 데이터를 관리할 수 있도록 해준다.
#numpy : 수학 및 과학연산(벡터연산)을 위한 파이썬 패키지로 다차원 배열 
#데이터 구조와 관련 도구들을 제공한다. 파이썬으로 수치해석, 통계 관련 작업을
#수행할 때 사용되는 다양한 기능들을 포함하고 있다.
#matplotlib : 데이터 분석 결과를 표현할 수 있는 강력한 시각화 도구이다.
#그래프를 통해 데이터를 한눈에 확인 하기 더욱 유용해진다.
```
