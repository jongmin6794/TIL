```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
tf.set_random_seed(777)
np.random.seed(777)
```

```python
#ydata=xdata*w+b
w=tf.Variable(tf.random_normal([1]),name='weight')
b=tf.Variable(tf.random_normal([1]),name='bias')
```

```python
x=tf.placeholder(tf.float32, shape=[None])
y=tf.placeholder(tf.float32, shape=[None])
#[None,3]
```

```python
hf=x*w+b
cost=tf.reduce_mean(tf.square(hf-y))
#예측값 - 실제값 -> 제곱 -> 합 -> 평균
```

```python
opt=tf.train.GradientDescentOptimizer(learning_rate=0.01)
train=opt.minimize(cost)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    _,cv,bv,wv=sess.run([train,cost,b,w],feed_dict={x:[1,2,3],y:[2,3,4]})
    if step%100==0:
        print(step, cv, bv, wv)
#y=1*x+1
```

```python
xdata=[[1,2],
       [2,3],
       [3,1],
       [4,3],
       [5,3],
       [6,2]]
ydata=[[0],
       [0],
       [0],
       [1],
       [1],
       [1]]
```

```python
x=tf.placeholder(tf.float32, shape=[None,2])
y=tf.placeholder(tf.float32, shape=[None,1])
```

```python
w=tf.Variable(tf.random_normal([2,1]),name="weight")
b=tf.Variable(tf.random_normal([1]),name="bias")
```

```python
hf=tf.sigmoid(tf.matmul(x,w)+b)
```

```python
cost=-tf.reduce_mean(y*tf.log(hf)+(1-y)*tf.log(1-hf))
```

```python
train=tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
#0.5 기준(크면->1, 작으면->0)
predicted=tf.cast(hf>0.5,dtype=tf.float32)
accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted,y),dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(20001):
        _,cv=sess.run([train,cost], feed_dict={x:xdata,y:ydata})
        if step%200==0:
            print(step,cv)
    hv,pv,av=sess.run([hf,predicted,accuracy],feed_dict={x:xdata,y:ydata})
    print("예측값:",hv, "예측분류:",pv,"정확도:",av)
```

```python
dt=pd.DataFrame(data)
dt=dt.set_index(0)
dt.index.names=[None]
dt # 470,17
```

```python
data[:][0].value_counts()
dt.iloc[:,-1].value_counts() # 0 : 400, 1 : 70
```

```python
scaler=StandardScaler()
x_data=scaler.fit_transform(dt.iloc[:,1:-1])
# x_data=np.array(dt.iloc[:,1:-1])
x_data.shape#470,15
y_data=np.array(dt.iloc[:,-1])
y_data=y_data.reshape(470,1)

x=tf.placeholder(tf.float32, shape=[None,15])
y=tf.placeholder(tf.float32, shape=[None,1])
w=tf.Variable(tf.random_normal([15,1]),name='weight')*5e-5
b=tf.Variable(tf.random_normal([1]),name='bias')*5e-5

hf=tf.sigmoid(tf.matmul(x,w)+b)
cost=-tf.reduce_mean(y*tf.log(hf)+(1-y)*tf.log(1-hf))

train=tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
predicted=tf.cast(hf>0.5,dtype=tf.float32)
accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted,y),dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(20001):
        _,cv=sess.run([train,cost],feed_dict={x:x_data,y:y_data})
        if step%200==0:
            print(step,cv)
    hv,pv,av=sess.run([hf,predicted,accuracy], feed_dict={x:x_data,y:y_data})
    print("예측값:",hv,"예측분류:",pv,"정확도:",av)
```
